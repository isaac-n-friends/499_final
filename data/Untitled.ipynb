```python
import matplotlib.pyplot as plt
```

```python
country = ['China', 'Iran', 'Saudi Arabia', 'Iraq','Egypt', 'United States', 'Pakistan','Somalia']
executions = [1000, 251,184, 100,32,22,14,12]
```

```python
plt.figure(figsize=(10,7))
plt.bar(country, executions)
plt.xlabel=('Country')
plt.ylabel=('Executions')
plt.xticks(rotation=45)
plt.title('Executions Per Country (2019)', fontsize=15)
plt.grid(axis='y', alpha=0.5)
plt.savefig('deaths.jpg')
```
